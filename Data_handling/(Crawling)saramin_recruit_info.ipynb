```python
# -*- coding: utf-8 -*-
# import sys
# reroad
# sys.setdefaultencoding('utf-8')

from bs4 import BeautifulSoup
import urllib.request
import pandas as pd
import datetime
```

```python
def public_recruit(result):
    pages_url = 'https://www.saramin.co.kr/zf_user/jobs/public/list/page/1?sort=ud&listType=public&public_list_flag=y#searchTitle'
    pageshtml = urllib.request.urlopen(pages_url)
    soupPage = BeautifulSoup(pageshtml, 'html.parser')
    pages = len(soupPage.select('a.page'))+1
    
    for page in range(1,pages+1):
        saramin_url = 'https://www.saramin.co.kr/zf_user/jobs/public/list/page/%d?sort=ud&listType=public&public_list_flag=y#searchTitle' %page
        print(saramin_url)
        html = urllib.request.urlopen(saramin_url)
        soupSaramin = BeautifulSoup(html, 'html.parser')
        #tag_tbody = soupSaramin.select('tbody')
        for corps in soupSaramin.select('tbody>tr'):
            if len(corps) <= 3:
                continue
            corp_td = corps.select('td')
            company_nm = corp_td[1].select('a[class=str_tit]')    #1.기업명 done
            if len(company_nm)==0:                                #
                company_nm=None                                  #
            else:                                                #
                company_nm=company_nm[0].string                      #
            job_title = corp_td[2].a.string                #2.제목 done
            career = corp_td[3].select_one("p.career").string        #4.경력 done
            education = corp_td[3].select_one("p.education").string        #5.학력 done 
            employment_type = corp_td[4].select("p[class=employment_type]")  #6.고용형태 done
            if len(employment_type) == 0:                                    #
                employment_type = None                                        #
            else:                                                            #
                employment_type = employment_type[0].string                 #
            work_place = corp_td[4].select_one("p.work_place").string    #7.회사위치 done
            deadlines = corp_td[5].select_one("p.deadlines").string        #8신청기한 done
            
            job_sector = corp_td[2].select("div.job_sector > span")    #3.직무 done
            sector = ""
            for i in corp_td[2].select("div.job_sector > span"):
                sector += " /" + i.string
        
            result.append([company_nm]+[job_title]+[sector]+[career]+[education]+[employment_type]+[work_place]+[deadlines])
    return
```

```python
result=[]
public_recruit(result)
result
```

```python
saramin_tbl = pd.DataFrame(result, columns = ('기업','공지','직무','경력','학력','채용형태','지역','신청기한'))
saramin_tbl.to_csv("C:/Users/pcw24/My_BigData/6장_data/saramin.csv", encoding = "utf-8-sig", mode = "w", index = True)
```
