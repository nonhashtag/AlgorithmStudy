```python
#데이터 수집을 위한 import
import numpy as np
import pandas as pd

from sklearn.datasets import load_boston
boston =  load_boston()
```

```python
#Training을 위한 import
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.metrics import mean_squared_error, r2_score
```

```python
#보스턴 데이터프레임 선언
boston_df = pd.DataFrame(boston.data, columns = boston.feature_names)
```

```python
boston_df["PRICE"] = boston.target
```

```python
boston_df["PRICE"] = boston.target
```

```python
boston_df
```

# Training 시작 (전체 요소)

```python
# X, Y 분할
Y = boston_df['PRICE']
X = boston_df.drop(['PRICE'], axis=1, inplace=False)
```

```python
#Train 데이터셋과 Test 데이터셋 분할
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.3, random_state=156)
```

```python
# 선형회귀분석 모델
lr = LinearRegression()
```

```python
# 선형회귀분석 모델 훈련
lr.fit(X_train, Y_train)
```

```python
# 선형회귀분석 : 평가 데이터에 대한 예측 수행 -> 예측 결과 Y_predict 구하기
Y_predict = lr.predict(X_test)
```

```python
mse = mean_squared_error(Y_test, Y_predict)
rmse = np.sqrt(mse)

print('MSE : {0:.3f}, RMSE : {1:.3f}'.format(mse, rmse))
print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test, Y_predict)))
```

```python
print('Y 절편 값: ', lr.intercept_)
print('회귀 계수 값: ', np.round(lr.coef_, 1))
```

```python
coef = pd.Series(data = np.round(lr.coef_, 2), index=X.columns)
coef.sort_values(ascending = False)
```

### 변수 선정의 유효성 검증단계

```python
boston_df2 = pd.DataFrame(boston.data, columns = boston.feature_names)
boston_df2["PRICE"] = boston.target
```

```python
lr3 = LinearRegression()
R2 = []
```

```python
for i in boston_df2:
    for j in boston_df2.drop([i,'PRICE'], axis=1, inplace=False):
        Y3 = boston_df2['PRICE']
        X3 = boston_df2[[i,j]]
        X_train3, X_test3, Y_train3, Y_test3 = train_test_split(X3, Y3, test_size=0.3, random_state=156)
        lr3.fit(X_train3, Y_train3)
        Y_predict3 = lr3.predict(X_test3)
        # 아래의 print문 2개의 주석을 해제하면, 각 2개의 컬럼별로의 R^2를 확인할 수 있다.
        #print('{} {}'.format(i,j), end=" : ")
        #print('R^2(Variance score) : {0:.3f}'.format(r2_score(Y_test3, Y_predict3)))
        R2.append(r2_score(Y_test3, Y_predict3))
    del boston_df2[i]
```

```python
max(R2)
# 모든 R^2의 값들에 대한 최댓값은 0.690(반올림값)이다.
```
